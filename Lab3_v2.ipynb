```python
kolory = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='DarkRed')
for i in ['Alcohol','Malic.acid','Ash','Acl','Mg','Phenols','Flavanoids',
          'Nonflavanoid.phenols','Proanth','Color.int','Hue','OD','Proline']:
    dane[i].plot.box(color=kolory)
    plt.savefig("Wykresy\BoxPlot_"+str(i)+".pdf")
    plt.show()
```

```python
palette1 = ['blue','orange','green']
for i in ['Alcohol','Malic.acid','Ash','Acl','Mg','Phenols','Flavanoids',
          'Nonflavanoid.phenols','Proanth','Color.int','Hue','OD']:
    for j in ['Malic.acid','Ash','Acl','Mg','Phenols','Flavanoids',
              'Nonflavanoid.phenols','Proanth','Color.int','Hue',
               'OD','Proline']:
        if i != j:
            sns.lmplot(x=i, y=j, data=dane, fit_reg=False, legend=True, hue='Wine', palette=palette1)
            plt.savefig("Wykresy\Lmplot_"+str(i)+str(j)+".pdf")
```

```python
plt.pie([dane[dane['Wine'] == 1]['Wine'].count(),dane[dane['Wine'] == 2]['Wine'].count(),
         dane[dane['Wine'] == 3]['Wine'].count()], 
        labels=['Klasa 1.','Klasa 2.', 'Klasa 3.'], autopct='%1.1f%%')
plt.title('Rozkład procentowy klas win')
plt.savefig("Wykresy\Klasy_win.pdf")
plt.show()
```
